```r
# 1. Definição dos Dados de Entrada ---------------------------------------

programas_nomes <- c("Saúde da Família", "Vigilância Genômica", "Educação em Saúde")
objetivos <- c("Ampliar cobertura da atenção primária",
               "Monitorar variantes de patógenos",
               "Capacitação de agentes comunitários")
publico_alvo <- c("População geral", "Laboratórios e Hospitais", "Servidores Públicos")
orcamentos <- c(1500000, 850000, 420000)
datas_inicio <- as.Date(c("2023-01-10", "2024-02-15", "2025-01-01"))
datas_fim <- as.Date(c("2026-12-31", "2024-12-20", "2025-12-31"))

# 2. Criar uma lista com as informações de cada programa ------------------

gestao_saude <- list(
  Programa1 = list(nome = programas_nomes[1], objetivo = objetivos[1], publico = publico_alvo[1], orcamento = orcamentos[1], inicio = datas_inicio[1], fim = datas_fim[1]),
  Programa2 = list(nome = programas_nomes[2], objetivo = objetivos[2], publico = publico_alvo[2], orcamento = orcamentos[2], inicio = datas_inicio[2], fim = datas_fim[2]),
  Programa3 = list(nome = programas_nomes[3], objetivo = objetivos[3], publico = publico_alvo[3], orcamento = orcamentos[3], inicio = datas_inicio[3], fim = datas_fim[3])
)

# 3. Calcular o orçamento total -------------------------------------------

orcamento_total <- sum(orcamentos)
cat("Orçamento Total Destinado: R$", format(orcamento_total, big.mark = ".", decimal.mark = ","), "\n")

# 4. Identificar o programa com o maior orçamento -------------------------

indice_maior <- which.max(orcamentos)
programa_maior_verba <- programas_nomes[indice_maior]
cat("Programa com maior orçamento:", programa_maior_verba, "(R$", orcamentos[indice_maior], ")\n")

# 5. Listar os programas em andamento atualmente --------------------------

data_hoje <- Sys.Date()
em_andamento <- programas_nomes[data_hoje >= datas_inicio & data_hoje <= datas_fim]

cat("Programas em andamento hoje (", format(data_hoje, "%d/%m/%Y"), "):\n")
print(em_andamento)

# BÔNUS: Visualização rápida em tabela (opcional)
df_programas <- data.frame(
  Nome = programas_nomes,
  Orcamento = orcamentos,
  Inicio = datas_inicio,
  Fim = datas_fim
)
View(df_programas)
```

```r
# 1. Entrada de Dados
notas <- c(8.5, 7.0, 9.2, 6.8, 8.0, 7.5, 9.0, 5.5, 8.8, 7.2)

# 2. Organização

# Ordenar em ordem crescente
notas_crescente <- sort(notas)
print("Notas em ordem crescente:")
print(notas_crescente)

# Inverter a ordem (da maior para a menor)
notas_decrescente <- rev(sort(notas))
print("Notas da maior para a menor:")
print(notas_decrescente)

# 3. Estatísticas Descritivas
media_notas <- mean(notas)
mediana_notas <- median(notas)
nota_maxima <- max(notas)
nota_minima <- min(notas)
desvio_padrao <- sd(notas)

# Exibindo os resultados formatados
cat("\n--- Resumo Estatístico ---\n")
cat("Média:", media_notas, "\n")
cat("Mediana:", mediana_notas, "\n")
cat("Nota mais alta:", nota_maxima, "\n")
cat("Nota mais baixa:", nota_minima, "\n")
cat("Desvio Padrão:", round(desvio_padrao, 2), "\n")

# Uso da função summary()
cat("\n--- Resumo Completo (Summary) ---\n")
print(summary(notas))

# 4. Frequência

tabela_frequencia <- table(notas)
print("Tabela de Frequência:")
print(tabela_frequencia)

# 5. Desafio Extra: Arredondamento e Comparação

notas_arredondadas <- round(notas, 1)

media_original <- mean(notas)
media_final <- mean(notas_arredondadas)

cat("\n--- Comparação de Médias ---\n")
cat("Média Original:", media_original, "\n")
cat("Média Arredondada:", media_final, "\n")
cat("Diferença:", media_original - media_final, "\n")
```

```r
Declarando UMA ÚNICA variável para todo o processo
# Ano escolhido: 2022
# UF escolhida: Espirito Santo

resultado <- fetch_datasus(
  year_start = 2022,
  year_end = 2022,
  uf = "ES",
  information_system = "SINASC"
) %>%
  process_sinasc() %>%
  mutate(
    # Extraindo o mês da data de nascimento
    mes_nascimento = month(DTNASC),
    # Criando nome do mês para melhor visualização
    nome_mes = factor(
      mes_nascimento,
      levels = 1:12,
      labels = c("Janeiro", "Fevereiro", "Março", "Abril", "Maio", "Junho",
                 "Julho", "Agosto", "Setembro", "Outubro", "Novembro", "Dezembro")
    )
  ) %>%
  group_by(mes_nascimento, nome_mes) %>%
  summarise(
    total_nascimentos = n(),
    .groups = "drop"
  ) %>%
  arrange(mes_nascimento)


# resultado
print(resultado)

install.packages("microdatasus")
install.packages("devtools")
```
